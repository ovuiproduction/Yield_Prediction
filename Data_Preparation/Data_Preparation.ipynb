```python
import pandas as pd
```

```python
df = pd.read_csv("Mango_dataset.csv")
```

```python
df.head()
```

```python
df['Reported Date'] = pd.to_datetime(df['Reported Date'], format='%d-%b-%y')
```

```python
df.head()
```

```python
df['Year'] = df['Reported Date'].dt.year
df['Month'] = df['Reported Date'].dt.month
```

```python
df.head()
```

```python
rainfall_data = pd.read_csv("Rainfall_data.csv")
```

```python
rainfall_data.head()
```

```python
rainfall_lookup = {(row['Year'], row['Month']): row['Rainfall (mm)'] for _, row in rainfall_data.iterrows()}

df['Rainfall(mm)'] = df.apply(
    lambda row: rainfall_lookup.get((row['Year'], row['Month']), row['Rainfall(mm)']),
    axis=1
)
```

```python
df.head()
```

```python
df.drop('Reported Date', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.to_csv('Mango_dataset_final.csv', index=False)
```
